```python
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline

f = open('res-retain.txt', 'r')
loss = []
for line in f:
    st = line.split(';')
    if len(st) > 2:
        loss_par = st[1].split(': ')[-1]
        loss.append(float(loss_par))
        
f.close()
print(len(loss))
plt.plot(loss)
```
